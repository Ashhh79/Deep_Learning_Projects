```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
breast_cance = datasets.load_breast_cancer()
```

```python
X = df.drop(['target'],1)
Y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

stdx_train = scaler.fit_transform(x_train)

stdx_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                        keras.layers.Flatten(input_shape = (30,)),
                        keras.layers.Dense(20, activation='relu'),
                        keras.layers.Dense(2,activation = 'sigmoid')

])
```

```python
from keras.src.engine.training import optimizer
model.compile(optimizer = 'adam',
          loss = 'sparse_categorical_crossentropy',
          metrics = ['accuracy'])
```

```python
history = model.fit(stdx_train,y_train,validation_split=0.1,epochs=10)
```

Visualizing Accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data', 'validation data'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data', 'validation data'])
```

Accuracy on test dada

```python
loss, accuracy = model.evaluate(stdx_test,y_test)
```

```python
y_pred = model.predict(stdx_test)
```

model.predict gives the probability of each class for each data point

Converting the prediction probabilty to class labels

```python
input = (12.31,16.52,79.19,470.9,0.09172,0.06829,0.03372,0.02272,0.172,0.05914,0.2505,1.025,1.74,19.68,0.004854,0.01819,0.01826,0.007965,0.01386,0.002304,14.11,23.21,89.71,611.1,0.1176,0.1843,0.1703,0.0866,0.2618,0.07609)

input_arr = np.asarray(input)

input_arr = input_arr.reshape(1,-1)

input_arr_std = scaler.transform(input_arr)

prediction = model.predict(input_arr_std)

print(prediction)

prediction_label = np.argmax(prediction)

print(prediction_label)

if prediction_label==0:
  print('Malignant')
else:
    print('Benign')
```
